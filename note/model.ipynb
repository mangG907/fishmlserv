```python
fig, ax = plt.subplots()

ax.scatter(bream_length, bream_weight, label = "Bream", edgecolor = 'black')
ax.scatter(smelt_length, smelt_weight, label = "Smelt", edgecolor = 'black')

plt.legend()
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
length = bream_length + smelt_length
weight = bream_weight + smelt_weight
import pandas as pd
df = pd.DataFrame({'length' : length, 'weight' : weight})
df.head()
```

```python
bream = [0] * 35
smelt = [1] * 14
target = bream + smelt
df['target'] = target
df.head()
```

```python
model = KNeighborsClassifier()
y = df.pop('target')
x = df
model.fit(x, y)
```

```python
import pickle 
with open('model.pkl', 'wb') as f:
    pickle.dump(model, f)
    
```
